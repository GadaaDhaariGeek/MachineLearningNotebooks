```python
# In this notebook we will talk about L1 regularization which is a penalization method, which penalizes weights which are 
# larger in number. It is also called shrinkage method as it shrinks down the weight associated with features having larger
# weights. 

# L1 regularization can also be used as feature extractor or feature selector because it can shrink the weight to 0.
```

```python
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
# number of points 
# what will happen for N = 100 ?
N = 50

# number of dimensions
D = 50

# generating random values between -5 to +5
X = (np.random.random((N, D)) - 0.5) * 10

# true weights for these features 
true_w = np.array([1, -.5, 3] + [0]*(D-3))

# Making Y using true weights and add some random noise to it
Y = X.dot(true_w) + np.random.rand(N)*.5
```

```python
# Performing gradient descent
plt.figure(figsize=(20, 8))
learning_rate = 1e-3
w = np.random.randn(D) / np.sqrt(D)
# l1 = 10
l1 = [0, 5, 10, 15]
for l in l1:
    costs = []
    for i in range(100):
        Yhat = X.dot(w)
        delta = Yhat - Y

        w = w - learning_rate*(X.T.dot(delta) + l*np.sign(w))

        mse = delta.dot(delta)/N
        costs.append(mse)
    plt.plot(costs, label= "l1 strength: " + str(l))
plt.legend()
plt.show()
```
