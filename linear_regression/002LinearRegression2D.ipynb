```python
import numpy as np

# To deal with warnings
import warnings 
warnings.filterwarnings("ignore")

# To plot 3d data
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
# Generating the 2D data
N = 200

# weights of 2 features of X
w = np.array([2, 3])

# Generating the values from uniform distribution of dimension (Nx2)
X = np.random.uniform(low=0, high=100, size=(N, 2))

# Generating y = w1X1 + w2X2 + 1 + gaussian noise
Y = np.dot(X, w) + 1 + np.random.normal(scale=5, size=N)
```

```python
# Printing shape of data 
X.shape, Y.shape
```

```python
# plotting figure in 3D 
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X[:, 0], X[:, 1], Y, edgecolors="black", color="green")
plt.show()
```

```python
# Lets plot with plotly to better visualize the data
fig = px.scatter_3d(x=X[:, 0], y=X[:, 1], z=Y)
fig.update_traces(marker=dict(size=6, line=dict(width=2, color='black'), color="green"), selector=dict(mode='markers'))
fig.show()
```

```python
# We can see data can be observed as a plane in 3d.
```

```python
# After solving the multiple linear regression we came up with the following solution 
# w = (X.T.X)^-1(X.T.Y)
w = np.linalg.solve(a = np.dot(X.T, X), b = np.dot(X.T, Y))
```

```python
# Priting the weights 
# We can see our model estimates the values of w1 and w2 very closely
print(w)
```

```python
# Calculating y values thorugh our model
Yhat = np.dot(a=X, b=w)
```

```python
# Determining how good the model is 
ss_res = (Y - Yhat).dot(Y - Yhat)
ss_tot = (Y - Y.mean()).dot(Y - Y.mean())
r2 = 1 - (ss_res/ss_tot)
print("R2 is: ", r2)
```

```python
# Again since our data is created accroding the linear relationship and the assumptions of residual, r squared value is 
# pretty good. We can also plot the residuals to see whether they follow the normal distribution or not 
sns.distplot((Y - Yhat), bins=15)
plt.show()
```

```python
# The residuals look like a normal distribution which is the assumption of linear regression.
```
