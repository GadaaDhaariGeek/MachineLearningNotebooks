```python
# In this notebook we will talk about L1 regularization which is a penalization method, which penalizes weights which are 
# larger in number. It is also called shrinkage method as it shrinks down the weight associated with features having larger
# weights. 

# L1 regularization can also be used as feature extractor or feature selector because it can shrink the weight to 0.
```

```python
# L1 regularization is used when D is greater than N or comparable to N, to do feature selection making few of the 
# coefficients of feature as 0 and helps achieve sparsity.

# Here we have Laplace distribution to use and have Laplace pior and solve for the weights.
```

```python
# Taking the derivative of the cost function and making it zero.
# The equation which comes is dellJ/dellW = -2X.T.Y + 2X.T.X.w + lambda*sign(w) = 0
# Since sign function return -1 for <0, 1 for >0 and 0 for 0, we cant have an inverse of this fucntion.
# So we use gradient descent and not the normal equation
```

```python
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
# number of points 
# what will happen for N = 100 ?
N = 50

# number of dimensions
D = 50

# generating random values between -5 to +5
X = (np.random.random((N, D)) - 0.5) * 10

# true weights for these features 
true_w = np.array([1, -.5, 3] + [0]*(D-3))

# Making Y using true weights and add some random noise to it
Y = X.dot(true_w) + np.random.rand(N)*.5
```

```python
# Performing gradient descent
plt.figure(figsize=(20, 8))
learning_rate = 1e-3
w = np.random.randn(D) / np.sqrt(D)
# l1 = 10
l1 = [0, 1, 2, 5]
for l in l1:
    costs = []
    for i in range(100):
        Yhat = X.dot(w)
        delta = Yhat - Y

        w = w - learning_rate*(X.T.dot(delta) + l*np.sign(w))

        mse = delta.dot(delta)/N
        costs.append(mse)
    plt.plot(costs, label= "l1 strength: " + str(l), linewidth=2)
plt.legend()
plt.show()
```

```python
# Performing gradient descent
plt.figure(figsize=(20, 8))
plt.plot(true_w, label="true weights")
learning_rate = 1e-3
w = np.random.randn(D) / np.sqrt(D)
# l1 = 10
l1 = [0, 1, 2, 5]
for l in l1:
    costs = []
    for i in range(100):
        Yhat = X.dot(w)
        delta = Yhat - Y

        w = w - learning_rate*(X.T.dot(delta) + l*np.sign(w))

        mse = delta.dot(delta)/N
        costs.append(mse)
    plt.plot(w, label= "l1 strength: " + str(l))
plt.legend()
plt.xticks(np.arange(0, 50))
plt.xlabel("Weights as indices    ----->")
plt.ylabel("Weight values for each feature   ---->")
plt.show()
```
