```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# L2 regularizaation method is use dto shrink the unimportant feature's coefficients and also diminishes the effect of 
# outliers
```

```python
N = 50

X = np.linspace(0, 10, N)
```

```python
Y = .5*X + np.random.randn(N)
```

```python
# making the outliers
Y[-1] += 30
Y[-2] += 30
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
X = np.vstack((np.ones(N), X)).T
```

```python
w_ml = np.linalg.solve(X.T.dot(X), X.T.dot(Y))
```

```python
Yhat_ml = X.dot(w_ml)
```

```python
plt.scatter(X[:, 1], Y)
plt.scatter(X[:, 1], Yhat_ml)
plt.show()
```

```python
# L2 regularization solution
l2 = 1000.0

# We add the term l2*np.eye(2) because we get the equation where lambda*I comes, when we try to solve the for the equation
w_map = np.linalg.solve(l2*np.eye(2) + X.T.dot(X), X.T.dot(Y))
```

```python
Yhat_map = X.dot(w_map)
```

```python
plt.scatter(X[:, 1], Y, color="red")
plt.plot(X[:, 1], Yhat_ml, label="maximum likelihood", color="green", linewidth=3)
plt.plot(X[:, 1], Yhat_map, label="map", linewidth=3)
plt.legend()
plt.show()
```
