```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
N = 100
D = 2

X = np.random.randn(N, D)

X[:50, :] = X[:50, :] - 2*np.ones((50, D))
X[50:, :] = X[50:, :] + 2*np.ones((50, D))

T = np.array([0]*50 + [1]*50)

ones = np.ones((N, 1))
Xb = np.concatenate((ones, X), axis=1)

# closed form solution
w = np.array([0, 4, 4])

z = Xb.dot(w)

y = sigmoid(z)
```

```python
# Plotting the data
plt.scatter(x = X[:, 0], y = X[:, 1], c=T, s=50, edgecolor="black", alpha = .5)

x_axis = np.linspace(-6, 6, 100)
y_axis = -x_axis
plt.plot(x_axis, y_axis)
plt.show()
```
