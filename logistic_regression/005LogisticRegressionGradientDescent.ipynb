```python
# Coin toss
# p(H) = p and p(T) = 1 - p
# N = 10, 7H and 3T

# then likelihood
# L = p^7(1-p)3
# we calculate log likelihood of this 

# and try to maximize that, then 
# p comes (7/10) which is probability for heads 
```

```python
# How to get optimized weights for logistic regression?
# We can take derivative of cross entropy error
# But we cant make it zero and solve for the weights like linear regression (why ?)

# So, we use gradient descent for minmizing the corss entropy error
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 100
D = 2

N_per_class = int(N/2)


X = np.random.randn(N,D)

# center the first 50 points at (-2,-2)
X[:N_per_class,:] = X[:N_per_class,:] - 2*np.ones((N_per_class,D))

# center the last 50 points at (2, 2)
X[N_per_class:,:] = X[N_per_class:,:] + 2*np.ones((N_per_class,D))

# labels: first N_per_class are 0, last N_per_class are 1
T = np.array([0]*N_per_class + [1]*N_per_class)

# add a column of ones
# ones = np.array([[1]*N]).T # old
ones = np.ones((N, 1))
Xb = np.concatenate((ones, X), axis=1)

# randomly initialize the weights
w = np.random.randn(D + 1)

# calculate the model output
z = Xb.dot(w)

def sigmoid(z):
    return 1/(1 + np.exp(-z))


Y = sigmoid(z)

# calculate the cross-entropy error
def cross_entropy(T, Y):
    E = 0
    for i in range(len(T)):
        if T[i] == 1:
            E -= np.log(Y[i])
        else:
            E -= np.log(1 - Y[i])
    return E
```

```python
# Doing gradient descent
learning_rate = .1
for i in range(1000):
    if i % 100 == 0:
        print(cross_entropy(T, Y))
        
    # gradient descent weights update 
    w -= learning_rate*Xb.T.dot(Y - T)
    
    Y = sigmoid(Xb.dot(w))
    

print("Final w: ", w)
```

```python
# Plotting the data
plt.scatter(X[:,0], X[:,1], c=T, s=80, alpha=0.5, edgecolors="black")
x_axis = np.linspace(-6, 6, 100)
y_axis = -(w[0] + x_axis*w[1]) / w[2]
plt.plot(x_axis, y_axis)
plt.show()
```
